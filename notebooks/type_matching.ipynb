```python
from pydantic import BaseModel
from typing import Type, TypeVar

class MyModel(BaseModel):
    name: str
    age: int
```

```python
model_name = MyModel.__name__
print(model_name)  # Output: 'MyModel'
```

```python
def get_model_class(model_name: str) -> Type[BaseModel]:
    # Assuming the model classes are in the current global namespace
    return globals().get(model_name, None)
```

```python
model_class = get_model_class(model_name)
```

```python
json_str = {
    "name": "Wassup",
    "age": 1
}

print(model_class.model_validate(json_str))
```
